```python
from __future__ import print_function
import sys
sys.path.append("..")
from util.util import lr_scheduler
from datasets.es_imagenet import ESImagenet_Dataset
import LIAF
from LIAFnet.LIAFResNet import *
import collections
import torchvision
import torchvision.transforms as transforms
import torch.distributed as dist 
import torch.nn as nn
import argparse, pickle, torch, time, os,sys
from importlib import import_module

import matplotlib.pyplot as plt # plt 用于显示图片
import matplotlib.image as mpimg # mpimg 用于读取图片
import matplotlib

import cv2,os
from skimage.metrics import structural_similarity as ssim

save_folder = './Visresult/ES_ImageNet_R18_Parameters_CNNbase_ADAM'
if not os.path.isdir(save_folder):
    os.mkdir(save_folder)
        
        
class Hungarian():
    def __init__(self, input_matrix=None, is_profit_matrix=False):
        if input_matrix is not None:
            my_matrix = np.array(input_matrix)
            self._input_matrix = np.array(input_matrix)
            self._maxColumn = my_matrix.shape[1]
            self._maxRow = my_matrix.shape[0]
            matrix_size = max(self._maxColumn, self._maxRow)
            pad_columns = matrix_size - self._maxRow
            pad_rows = matrix_size - self._maxColumn
            my_matrix = np.pad(my_matrix, ((0,pad_columns),(0,pad_rows)), 'constant', constant_values=(0))
            if is_profit_matrix:
                my_matrix = self.make_cost_matrix(my_matrix)
            self._cost_matrix = my_matrix
            self._size = len(my_matrix)
            self._shape = my_matrix.shape
            self._results = []
            self._totalPotential = 0
        else:
            self._cost_matrix = None
    def make_cost_matrix(self,profit_matrix):
        matrix_shape = profit_matrix.shape
        offset_matrix = np.ones(matrix_shape, dtype=int) * profit_matrix.max()
        cost_matrix = offset_matrix - profit_matrix
        return cost_matrix
    def get_results(self):
        return self._results
    def calculate(self):
        result_matrix = self._cost_matrix.copy()
        # 1
        for index, row in enumerate(result_matrix):
            result_matrix[index] -= row.min()
        # 2
        for index, column in enumerate(result_matrix.T):
            result_matrix[:, index] -= column.min()
        total_covered = 0
        while total_covered < self._size:
            cover_zeros = CoverZeros(result_matrix)
            single_zero_pos_list = cover_zeros.calculate()
            covered_rows = cover_zeros.get_covered_rows()
            covered_columns = cover_zeros.get_covered_columns()
            total_covered = len(covered_rows) + len(covered_columns)
            if total_covered < self._size:
                result_matrix = self._adjust_matrix_by_min_uncovered_num(result_matrix, covered_rows, covered_columns)
        self._results = single_zero_pos_list
        value = 0
        for row, column in single_zero_pos_list:
            value += self._input_matrix[row, column]
        self._totalPotential = value

    def get_total_potential(self):
        return self._totalPotential

    def _adjust_matrix_by_min_uncovered_num(self, result_matrix, covered_rows, covered_columns):
        adjusted_matrix = result_matrix
        elements = []
        for row_index, row in enumerate(result_matrix):
            if row_index not in covered_rows:
                for index, element in enumerate(row):
                    if index not in covered_columns:
                        elements.append(element)
        min_uncovered_num = min(elements)
        for row_index, row in enumerate(result_matrix):
            if row_index not in covered_rows:
                for index, element in enumerate(row):
                    if index not in covered_columns:
                        adjusted_matrix[row_index,index] -= min_uncovered_num
        for row_ in covered_rows:
            for col_ in covered_columns:
                adjusted_matrix[row_,col_] += min_uncovered_num
        return adjusted_matrix


class CoverZeros():

    def __init__(self, matrix):

        self._zero_locations = (matrix == 0)
        self._zero_locations_copy = self._zero_locations.copy()
        self._shape = matrix.shape
        self._covered_rows = []
        self._covered_columns = []

    def get_covered_rows(self):
        return self._covered_rows

    def get_covered_columns(self):
        return self._covered_columns

    def row_scan(self,marked_zeros):
        min_row_zero_nums = [9999999,-1]
        for index, row in enumerate(self._zero_locations_copy):#index为行号
            row_zero_nums = collections.Counter(row)[True]
            if row_zero_nums < min_row_zero_nums[0] and row_zero_nums!=0:
                min_row_zero_nums = [row_zero_nums,index]
        row_min = self._zero_locations_copy[min_row_zero_nums[1],:]
        row_indices, = np.where(row_min)
        marked_zeros.append((min_row_zero_nums[1],row_indices[0]))
        self._zero_locations_copy[:,row_indices[0]] = np.array([False for _ in range(self._shape[0])])
        self._zero_locations_copy[min_row_zero_nums[1],:] = np.array([False for _ in range(self._shape[0])])

    def calculate(self):
        ticked_row   = []
        ticked_col   = []
        marked_zeros = []
        while True:
            if True not in self._zero_locations_copy:
                break
            self.row_scan(marked_zeros)
        independent_zero_row_list = [pos[0] for pos in marked_zeros]
        ticked_row = list(set(range(self._shape[0])) - set(independent_zero_row_list))
        TICK_FLAG = True
        while TICK_FLAG:
            TICK_FLAG = False
            for row in ticked_row:
                row_array = self._zero_locations[row,:]
                for i in range(len(row_array)):
                    if row_array[i] == True and i not in ticked_col:
                        ticked_col.append(i)
                        TICK_FLAG = True
            for row,col in marked_zeros:
                if col in ticked_col and row not in ticked_row:
                    ticked_row.append(row)
                    FLAG = True
        self._covered_rows    = list(set(range(self._shape[0])) - set(ticked_row))
        self._covered_columns = ticked_col
        return marked_zeros


def norm_canve(canve):
    return (canve - torch.min(canve))/(torch.max(canve)-torch.min(canve)) * 255



device = torch.device("cpu")
batch_size = 1
SSIM_list_liaf = []
SSIM_list_lif = []
wieght_size = 0
dim = 3
acc_cnn = [19,32,46,50]            
acc_lif = [5,16,25,31]
acc_liaf = [19,39,46,51]

#base model
cnn_dict_base = {}
checkpoint = torch.load('../exp_ES_imagenet/ResNet18_imagenet_CNN/50.pkl', map_location=torch.device('cpu'))
modules = import_module('LIAFnet.LIAFResNet_18')
CNN_config  = modules.Config()
CNN_config.cfgCnn = [3, 64, 7, True]
CNN = LIAFResNet(CNN_config)
CNN.load_state_dict(checkpoint)
CNNconv_base = CNN.conv1

for _,param in enumerate(CNNconv_base.named_parameters()):
    if param[0] == 'kernel.weight':
        wieght_size = param[1].size(3)
        print(param[1].size())
        
for _,param in enumerate(CNNconv_base.named_parameters()):
    if param[0] == 'kernel.weight':
        for i in range(64):
            cnn_dict_base[i] = param[1][i,:,:,:]

for model_index in range(len(acc_lif)):
    
    model_lif = acc_lif[model_index]
    model_liaf = acc_liaf[model_index]
    model_cnn  = acc_cnn[model_index]
    
    cnn_dict = {}
    lif_dict = {}
    liaf_dict = {}
    pair_dict_lif = {}
    pair_dict_liaf = {}
    
    checkpoint = torch.load('../exp_ES_imagenet/ResNet18_imagenet_CNN/'+str(model_cnn)+'.pkl', map_location=torch.device('cpu'))
    modules = import_module('LIAFnet.LIAFResNet_18')
    CNN_config  = modules.Config()
    CNN_config.cfgCnn = [3, 64, 7, True]
    CNN = LIAFResNet(CNN_config)
    CNN.load_state_dict(checkpoint)
    CNNconv = CNN.conv1
    for _,param in enumerate(CNNconv.named_parameters()):
        if param[0] == 'kernel.weight':
            a =  param[1]
            for i in range(64):
                cnn_dict[i] = param[1][i,:,:,:]
            
    checkpoint = torch.load('../exp_ES_imagenet/ResNet18_imagenet_LIF/'+str(model_lif)+'.pkl', map_location=torch.device('cpu'))
    LIF18_config  = modules.Config()
    LIF18_config.cfgCnn = [dim, 64, 7, True]
    LIF18_config.actfun = LIAF.LIFactFun.apply
    LIF18 = LIAFResNet(LIF18_config)
    LIF18.load_state_dict(checkpoint)
    LIFconv = LIF18.conv1   
    for _,param in enumerate(LIFconv.named_parameters()):
        if param[0] == 'kernel.weight':
            b =  param[1]
            for i in range(64):
                zero_canve = torch.zeros([3,wieght_size,wieght_size])
                zero_canve[0:dim,:,:] = param[1][i,:,:,:]
                lif_dict[i] = zero_canve
                
    
    checkpoint = torch.load('../exp_ES_imagenet/ResNet18_imagenet_LIAF/sgd_'+str(model_liaf)+'.pkl', map_location=torch.device('cpu'))
    LIAF18_config  = modules.Config()
    LIAF18_config.cfgCnn = [dim, 64, 7, True]
    LIAF18 = LIAFResNet(LIF18_config)
    LIAF18.load_state_dict(checkpoint)
    LIAFconv = LIAF18.conv1   
    for _,param in enumerate(LIAFconv.named_parameters()):
        if param[0] == 'kernel.weight':
            c =  param[1]
            for i in range(64):
                zero_canve = torch.zeros([3,wieght_size,wieght_size])
                zero_canve[0:dim,:,:] = param[1][i,:,:,:]
                liaf_dict[i] = zero_canve
    
    lif_profitmatrx = np.zeros([64,64])
    liaf_profitmatrx = np.zeros([64,64])
    
    gap = int(wieght_size*8.5)
    canve = torch.zeros(wieght_size*8,wieght_size*8+gap*2,3)
    
    for i in range(64):#LIF
        for j in range(64):#CNN
            img1 = lif_dict[i].permute(1,2,0).detach().numpy()
            img2 = cnn_dict_base[j].permute(1,2,0).detach().numpy()
            img1=cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
            img2=cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
            lif_profitmatrx[i,j] = 1-ssim(img1,img2)
            
    for i in range(64):#LIF
        for j in range(64):#CNN
            img1 = liaf_dict[i].permute(1,2,0).detach().numpy()
            img2 = cnn_dict_base[j].permute(1,2,0).detach().numpy()
            img1=cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
            img2=cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
            liaf_profitmatrx[i,j] = 1-ssim(img1,img2)   
            
    hungarian1 = Hungarian(lif_profitmatrx)
    hungarian1.calculate()
    print('calculating LIF...')
    print("Calculated value:\t", hungarian1.get_total_potential())  # = 12
    print("Results:\n\t", hungarian1.get_results())
    pair_list1 = hungarian1.get_results()
    SSIM_list_lif.append(hungarian1.get_total_potential())
    
    hungarian2 = Hungarian(liaf_profitmatrx)
    hungarian2.calculate()
    print('calculating LIAF...')
    print("Calculated value:\t", hungarian2.get_total_potential())  # = 12
    print("Results:\n\t", hungarian2.get_results())
    pair_list2 = hungarian2.get_results()
    SSIM_list_liaf.append(hungarian2.get_total_potential())
    
    print("-" * 80)   

    for pair in pair_list1:
        pair_dict_lif[pair[1]] = pair[0]
    for pair in pair_list2:
        pair_dict_liaf[pair[1]] = pair[0]
    
    canve = torch.ones(wieght_size*8,wieght_size*8+gap*2,3)*255

    for i in range(64):
        canve[i//8*wieght_size:(i//8+1)*wieght_size,i%8*wieght_size:(i%8+1)*wieght_size,0] = cnn_dict[i][0,:,:]
        canve[i//8*wieght_size:(i//8+1)*wieght_size,i%8*wieght_size:(i%8+1)*wieght_size,1] = cnn_dict[i][1,:,:]
        canve[i//8*wieght_size:(i//8+1)*wieght_size,i%8*wieght_size:(i%8+1)*wieght_size,2] = cnn_dict[i][2,:,:]

    for i in range(64):
        canve[i//8*wieght_size:(i//8+1)*wieght_size,gap+i%8*wieght_size:gap+(i%8+1)*wieght_size,1] = liaf_dict[pair_dict_liaf[i]][0,:,:]
        canve[i//8*wieght_size:(i//8+1)*wieght_size,gap+i%8*wieght_size:gap+(i%8+1)*wieght_size,2] = liaf_dict[pair_dict_liaf[i]][1,:,:]
        canve[i//8*wieght_size:(i//8+1)*wieght_size,gap+i%8*wieght_size:gap+(i%8+1)*wieght_size,0] = liaf_dict[pair_dict_liaf[i]][2,:,:]

    for i in range(64):
        canve[i//8*wieght_size:(i//8+1)*wieght_size,gap*2+i%8*wieght_size:gap*2+(i%8+1)*wieght_size,1] = lif_dict[pair_dict_lif[i]][0,:,:]
        canve[i//8*wieght_size:(i//8+1)*wieght_size,gap*2+i%8*wieght_size:gap*2+(i%8+1)*wieght_size,2] = lif_dict[pair_dict_lif[i]][1,:,:]
        canve[i//8*wieght_size:(i//8+1)*wieght_size,gap*2+i%8*wieght_size:gap*2+(i%8+1)*wieght_size,0] = lif_dict[pair_dict_lif[i]][2,:,:]

       
    canve[:,0:wieght_size*8,:] =  norm_canve(canve[:,0:wieght_size*8,:])
    canve[:,gap:gap+wieght_size*8,:] =  norm_canve(canve[:,gap:gap+wieght_size*8,:])
    canve[:,gap*2:gap*2+wieght_size*8,:] =  norm_canve(canve[:,gap*2:gap*2+wieght_size*8,:])

    task = save_folder+'/'+ str(model_index)+'.png'

    plt.rcParams['figure.figsize'] = (32.5,10) # 设置figure_size尺寸
    plt.rcParams['image.interpolation'] = 'nearest' # 设置 interpolation style
    plt.rcParams['savefig.dpi'] = 300 #图片像素
    plt.rcParams['figure.dpi'] = 300 #分辨率

    plt.figure("test") # 图像窗口名称
    plt.imshow(canve.int().numpy())
    plt.axis('off') # 关掉坐标轴为 off
    plt.savefig(task, format='png')
    plt.show()

    #print(torch.sum(abs(c-a))/64)
```
